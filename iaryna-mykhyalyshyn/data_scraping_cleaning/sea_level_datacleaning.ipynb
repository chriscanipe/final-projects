```python
import os
import glob
import pandas as pd
import numpy as np

path = "/Users/iaryna/Documents/Mizzou/AppDEV/final_project/line_final/data_scraping_cleaning/all_30years"
allfiles = glob.glob(os.path.join(path,"*.csv"))
count = 1



for file_ in allfiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    df.columns = ['year', 'month', 'day', 'level'] #add column names
    df = df[(df.year >= 1984) & (df.year < 2015)] #filter 50 years
    
    name_string = file_.split('/') #add city name
    city_name = name_string[10].replace("_", " ")
    city_name = city_name.replace(".csv", "")
#     count +=1

    df = df.groupby('year') #group by year
    df = df["level"].mean().round(1).reset_index().rename(columns={'level': 'level_norm'})
    starting_value = df['level_norm'].iloc[0] #get starting value
    


    df["level_norm"] = df.level_norm - starting_value
    df["city"] = city_name
#     df["count"] = count
    
#     df.index = df.index + 1 #start index from 1 for heatmap
    df.to_csv(file_, index=False)
    
    
    # get list of names
    last_value = df.level_norm.iloc[-1]
    print city_name,",", last_value 
```
